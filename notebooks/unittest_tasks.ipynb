```python
def calculate_discount(price,price_type):
    if price <= 0:
        raise ValueError("Invalid Price")
    elif price_type == 'premium':
        return price * 0.9
    return price
```

```python
import unittest

class TestCalculate(unittest.TestCase):
    def test_calculate_discount_invalid_price(self):
        price = 0
        price_type = 'premium'
        with self.assertRaises(ValueError) as context:
            calculate_discount(price,price_type)
        self.assertEqual(context.exception,"Invalid Price")

if(__name__ == "main"):
    unittest.main()
```
