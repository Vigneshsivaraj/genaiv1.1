Task: List vs Tuple â€“ Mutability & Behavior

Problem Statement:

Write a Python function that:

Accepts one list and one tuple

Tries to modify both inside the function

Prints the values inside the function

Prints the values outside the function

Clearly demonstrates why one changes and the other doesnâ€™t

```python
numbersList = list(range(10))
numbersTuple = tuple(numbersList)

def modify_data(numbersList,numbersTuple):
    numbersList[3] = 1
    print("Inside Function List",numbersList)
    try:
        numbersTuple[3] = 1
    except TypeError as er:
        print("Inside Function: You cannot assign value to exisitng tuple",er)
modify_data(numbersList,numbersTuple)

print("Outside function (list):", numbersList)
print("Outside function (tuple):", numbersTuple)
```

2.ðŸ”¹ Problem Statement

You are given a nested list representing users and their scores.

Your task is to:

Create a copy of the original list using a shallow copy

Modify the inner list in the copied version

Show that the original list is unexpectedly modified

Fix the issue using a deep copy

Clearly demonstrate the difference using prints

```python
nested_list = [1,2,[1,2,3,[1,2,3,4]]]

'''shallow copy'''

shallowedList = nested_list.copy()

shallowedList[1] = 3
shallowedList[2][3][1] = 1

print("shallowed List",shallowedList)
print("original list",nested_list)

'''Deep copy'''
deepCopiedList = []

def deepCopy(lists):
    if(len(lists) == 1):
        return lists
    result = []
    for item in lists:
        if(type(item) == list):
            result.append(deepCopy(item))
        else:
            result.append(item)
    return result

deepCopiedList = deepCopy(nested_list)

deepCopiedList[2][3][1] = 5
print("deepCopiedList",deepCopiedList)
print("original list",nested_list)
```

```python
nums = [1, 2, 3, 4, 5, 6]
i = len(nums) - 1

while i >= 0:
    if(nums[i] % 2 == 0):
        nums.pop(i)
    i = i - 1

print(nums)
```

ðŸ§  Task 9: Dictionary Iteration Patterns (Filter, Sort, Transform)

This task checks whether youâ€™re comfortable with real-world dictionary usage, not just looping.

ðŸ”¹ Problem Statement

Given this dictionary:

scores = {
    "alice": 85,
    "bob": 72,
    "charlie": 90,
    "david": 60
}


Your tasks:

Iterate and print all keyâ€“value pairs

Filter and keep only users with score â‰¥ 80

Sort the dictionary by score (descending)

Transform it into a list of strings like:

"alice scored 85"

```python
scores = {
    "alice": 85,
    "bob": 72,
    "charlie": 90,
    "david": 60
}

for key,value in scores.items():
    print(f"Key: {key} Value: {value}")

filtered_scores = {k:v for k,v in scores.items() if v >= 80}
sorted_scores = dict(sorted(filtered_scores.items(),key = lambda item: item[1],reverse=True))

result = [f"{k} in {v}" for k,v in sorted_scores.items() if v >= 90]
print(result)
```
