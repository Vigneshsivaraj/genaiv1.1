```python
#scikit-learn

import numpy as np
from sklearn.linear_model import LinearRegression

rng = np.random.RandomState(0)
X = rng.rand(100,1) * 10
y = 3*X.squeeze() + rng.randn(100)*2

model = LinearRegression().fit(X,y)
print(model.predict([[2]]))

```
